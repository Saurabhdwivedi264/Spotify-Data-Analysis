```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Read the dataset (Tracks)

df_tracks=pd.read_csv('C:/Users/Lenovo/Documents/Jupyter/tracks.csv')
df_tracks.head()
```

```python
# Check Null Values in the dataset

pd.isnull(df_tracks).sum()
```

```python
#Total values, column-wise in our dataset

df_tracks.info()
```

```python
# 10 Least Popular songs in our dataset

sorted_df=df_tracks.sort_values('popularity',ascending=True).head(10)
sorted_df
```

```python
# Statistical Values of our dataset

df_tracks.describe().transpose()
```

```python
# 10 Most Popular Songs, with Popularity greater than 90 in our dataset

most_popular=df_tracks.query('popularity>90',inplace=False).sort_values('popularity',ascending=False)
most_popular[:10]
```

```python
# Arrangement of the songs in order of their release date

df_tracks.set_index('release_date',inplace=True)
df_tracks.index=pd.to_datetime(df_tracks.index)
df_tracks.head()
```

```python
# Check the artist at 18th row (say)

df_tracks[["artists"]].iloc[18]
```

```python
# Convert the songs durations from milisecond to seconds in the dataset

df_tracks['duration']=df_tracks["duration_ms"].apply(lambda x: round(x/1000))
df_tracks.drop("duration_ms",inplace=True,axis=1)
```

```python
# Print the resulting data in seconds

df_tracks.duration.head()
```

```python
# Visualization : Co-Relation Map

corr_df=df_tracks.drop(['key','mode','explicit'],axis=1).corr(method='pearson')
plt.figure(figsize=(14,6))
heatmap=sns.heatmap(corr_df,annot=True,fmt='.1g',vmin=-1,center=0,cmap='inferno',linewidth=1,linecolor='Black')
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(),rotation=90)
```

```python
# Sampling from the original dataset to 0.4 percent

sample_df=df_tracks.sample(int(0.004*len(df_tracks)))
print(len(sample_df))
```

```python
# Correlation Graph using Regression (Loudness V/S Energy)

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness V/S Energy Correlation')
```

```python
#Correlation Graph using Regression (Popularity V/S Acousticness)

plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b').set(title='Popularity V/S Acousticness Correlation')
```

```python
# Create new column for release date

df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
# Number Of Songs released Per Year

sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title="Number Of Songs per Year")
```

```python
# Duration Of Songs

total_dr=df_tracks.duration
fig_dims=(18,7)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="Year V/S Duration")
plt.xticks(rotation=90)
```

```python
# Line plot to analyze year V/S duration of songs 

total_dr=df_tracks.duration
sns.set_style(style='whitegrid')
fig_dims=(10,5)
fig,ax=plt.subplots(figsize=fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year V/S Duration")
plt.xticks(rotation=60)
```

```python
# Read another dataset related to Spotify Features

df_genre=pd.read_csv("C:/Users/Lenovo/Documents/Jupyter/SpotifyFeatures.csv")
```

```python
# Print the first 5 rows of this dataset

df_genre.head()
```

```python
# Duration of the songs for different Genres

plt.title("Duration Of The Songs In Different Genres")
sns.color_palette('rocket',as_cmap=True)
sns.barplot(y='genre',x='duration_ms',data=df_genre)
plt.xlabel("Duration in ms")
plt.ylabel("Genres")
```

```python
# Top 5 Genres according to their popularity

sns.set_style(style='darkgrid')
plt.figure(figsize=(10,5))
famous=df_genre.sort_values("popularity",ascending=False).head(10)
sns.barplot(y="genre",x='popularity',data=famous).set(title="Top 5 Genres By Popularity")
```
